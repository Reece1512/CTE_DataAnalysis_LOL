## 2. Data processing

```python
import itertools

df = df[df["duration"] >= 900].copy()
teams = df.groupby(["game_id", "win"])["champion_name"].apply(list).reset_index()

def win_combos(teams, size):
    cnt = []
    for i in teams:
        cnt.extend(list(itertools.combinations(sorted(i), size)))
    return cnt

all_combos = win_combos(teams["champion_name"],2)
win_combos = win_combos(teams[teams["win"] == True]["champion_name"],2)

coms_cnt = pd.Series(all_combos).value_counts().rename("total_games")
win_coms_cnt = pd.Series(win_combos).value_counts().rename("wins")

duo_df = pd.concat([coms_cnt, win_coms_cnt], axis = 1).fillna(0)
duo_df["win_rate"] = np.where(duo_df["total_games"] > 0, duo_df["wins"]/duo_df["total_games"], 0)


top_duos = duo_df[duo_df["total_games"] >= 30].sort_values(by = ["win_rate", "total_games"], ascending = [False, False])
top_duos.head()
```
